```python
import pandas as pd
import numpy as np
```

```python
zone_map = {
    "FR": "France",
    "BE": "Belgium",
    "CH": "Switzerland",
    "DE-LU": "Germany",
    "DE-AT-LU": "Germany",
    "DE(TransnetBW)": "Germany",
    "DE(Amprion)": "Germany",
    "ES": "Spain",
    "GB": "Great Britain",
    "GB(ElecLink)": "Great Britain",
    "GB(IFA)": "Great Britain",
    "GB(IFA2)": "Great Britain",
    "IT-North": "Italy-North",
    "IT-North-FR": "Italy-North",
    "IT": "Italy-North",
    "LU": "Germany",
}
```

## 1. Prix day-ahead

```python
# (chemin_fichier, nom_pays, devise)
# devise : "EUR" ou "GBP"
price_files = [
    # Belgium
    ("data/energy_prices/Energy prices Belgium 2016.csv", "Belgium", "EUR"),
    ("data/energy_prices/Energy prices Belgium 2017.csv", "Belgium", "EUR"),
    ("data/energy_prices/Energy prices Belgium 2018.csv", "Belgium", "EUR"),
    ("data/energy_prices/Energy prices Belgium 2019.csv", "Belgium", "EUR"),
    ("data/energy_prices/Energy prices Belgium 2020.csv", "Belgium", "EUR"),
    ("data/energy_prices/Energy prices Belgium 2021.csv", "Belgium", "EUR"),
    ("data/energy_prices/Energy prices Belgium 2022.csv", "Belgium", "EUR"),
    ("data/energy_prices/Energy prices Belgium 2023.csv", "Belgium", "EUR"),
    ("data/energy_prices/Energy prices Belgium 2024.csv", "Belgium", "EUR"),
    ("data/energy_prices/Energy prices Belgium 2025.csv", "Belgium", "EUR"),
    # France
    ("data/energy_prices/Energy prices France 2016.csv", "France", "EUR"),
    ("data/energy_prices/Energy prices France 2017.csv", "France", "EUR"),
    ("data/energy_prices/Energy prices France 2018.csv", "France", "EUR"),
    ("data/energy_prices/Energy prices France 2019.csv", "France", "EUR"),
    ("data/energy_prices/Energy prices France 2020.csv", "France", "EUR"),
    ("data/energy_prices/Energy prices France 2021.csv", "France", "EUR"),
    ("data/energy_prices/Energy prices France 2022.csv", "France", "EUR"),
    ("data/energy_prices/Energy prices France 2023.csv", "France", "EUR"),
    ("data/energy_prices/Energy prices France 2024.csv", "France", "EUR"),
    ("data/energy_prices/Energy prices France 2025.csv", "France", "EUR"),
    # Germany
    ("data/energy_prices/Energy prices Germany 2019.csv", "Germany", "EUR"),
    ("data/energy_prices/Energy prices Germany 2020.csv", "Germany", "EUR"),
    ("data/energy_prices/Energy prices Germany 2021.csv", "Germany", "EUR"),
    ("data/energy_prices/Energy prices Germany 2022.csv", "Germany", "EUR"),
    ("data/energy_prices/Energy prices Germany 2023.csv", "Germany", "EUR"),
    ("data/energy_prices/Energy prices Germany 2024.csv", "Germany", "EUR"),
    ("data/energy_prices/Energy prices Germany 2025.csv", "Germany", "EUR"),
    # Great Britain (GBP)
    ("data/energy_prices/Energy prices Great Britain 2015.csv", "Great Britain", "GBP"),
    ("data/energy_prices/Energy prices Great Britain 2016.csv", "Great Britain", "GBP"),
    ("data/energy_prices/Energy prices Great Britain 2017.csv", "Great Britain", "GBP"),
    ("data/energy_prices/Energy prices Great Britain 2018.csv", "Great Britain", "GBP"),
    ("data/energy_prices/Energy prices Great Britain 2019.csv", "Great Britain", "GBP"),
    ("data/energy_prices/Energy prices Great Britain 2020.csv", "Great Britain", "GBP"),
    # Italy-North
    ("data/energy_prices/Energy prices Italy-North 2015.csv", "Italy-North", "EUR"),
    ("data/energy_prices/Energy prices Italy-North 2016.csv", "Italy-North", "EUR"),
    ("data/energy_prices/Energy prices Italy-North 2017.csv", "Italy-North", "EUR"),
    ("data/energy_prices/Energy prices Italy-North 2018.csv", "Italy-North", "EUR"),
    ("data/energy_prices/Energy prices Italy-North 2019.csv", "Italy-North", "EUR"),
    ("data/energy_prices/Energy prices Italy-North 2020.csv", "Italy-North", "EUR"),
    ("data/energy_prices/Energy prices Italy-North 2021.csv", "Italy-North", "EUR"),
    ("data/energy_prices/Energy prices Italy-North 2022.csv", "Italy-North", "EUR"),
    ("data/energy_prices/Energy prices Italy-North 2023.csv", "Italy-North", "EUR"),
    ("data/energy_prices/Energy prices Italy-North 2024.csv", "Italy-North", "EUR"),
    ("data/energy_prices/Energy prices Italy-North 2025.csv", "Italy-North", "EUR"),
    # Spain
    ("data/energy_prices/Energy prices Spain 2015.csv", "Spain", "EUR"),
    ("data/energy_prices/Energy prices Spain 2016.csv", "Spain", "EUR"),
    ("data/energy_prices/Energy prices Spain 2017.csv", "Spain", "EUR"),
    ("data/energy_prices/Energy prices Spain 2018.csv", "Spain", "EUR"),
    ("data/energy_prices/Energy prices Spain 2019.csv", "Spain", "EUR"),
    ("data/energy_prices/Energy prices Spain 2020.csv", "Spain", "EUR"),
    ("data/energy_prices/Energy prices Spain 2021.csv", "Spain", "EUR"),
    ("data/energy_prices/Energy prices Spain 2022.csv", "Spain", "EUR"),
    ("data/energy_prices/Energy prices Spain 2023.csv", "Spain", "EUR"),
    ("data/energy_prices/Energy prices Spain 2024.csv", "Spain", "EUR"),
    ("data/energy_prices/Energy prices Spain 2025.csv", "Spain", "EUR"),
    # Switzerland
    ("data/energy_prices/Energy prices Switzerland 2015.csv", "Switzerland", "EUR"),
    ("data/energy_prices/Energy prices Switzerland 2016.csv", "Switzerland", "EUR"),
    ("data/energy_prices/Energy prices Switzerland 2017.csv", "Switzerland", "EUR"),
    ("data/energy_prices/Energy prices Switzerland 2018.csv", "Switzerland", "EUR"),
    ("data/energy_prices/Energy prices Switzerland 2019.csv", "Switzerland", "EUR"),
    ("data/energy_prices/Energy prices Switzerland 2020.csv", "Switzerland", "EUR"),
    ("data/energy_prices/Energy prices Switzerland 2021.csv", "Switzerland", "EUR"),
    ("data/energy_prices/Energy prices Switzerland 2022.csv", "Switzerland", "EUR"),
    ("data/energy_prices/Energy prices Switzerland 2023.csv", "Switzerland", "EUR"),
    ("data/energy_prices/Energy prices Switzerland 2024.csv", "Switzerland", "EUR"),
    ("data/energy_prices/Energy prices Switzerland 2025.csv", "Switzerland", "EUR"),
]

prices_list = []

for path, country, currency in price_files:
    df = pd.read_csv(path)
    print(f"Lu : {path}  →  {len(df)} lignes")

    if "Sequence" in df.columns:
        df = df[df["Sequence"].isin(["Without Sequence", "Sequence 1"])]

    df["datetime"] = pd.to_datetime(
        df["MTU (CET/CEST)"].str[:19],
        format="%d/%m/%Y %H:%M:%S"
    )

    price_col = "Day-ahead Price (GBP/MWh)" if currency == "GBP" else "Day-ahead Price (EUR/MWh)"
    df = df.rename(columns={price_col: "price"})[["datetime", "price"]]
    df["country"] = country

    # Données quart-horaires (Allemagne) → horaire
    df["datetime"] = df["datetime"].dt.floor("h")
    df = df.groupby(["country", "datetime"], as_index=False).agg(price=("price", "mean"))

    prices_list.append(df)

prices = pd.concat(prices_list, ignore_index=True)
print(f"\nTotal : {len(prices):,} lignes — {prices['country'].nunique()} pays")
prices.head()
```

## 2. Flux physiques transfrontaliers

```python
flow_files = [
    "data/flows/Cross-border Physical Flows France 2014.csv",
    "data/flows/Cross-border Physical Flows France 2015.csv",
    "data/flows/Cross-border Physical Flows France 2016.csv",
    "data/flows/Cross-border Physical Flows France 2017.csv",
    "data/flows/Cross-border Physical Flows France 2018.csv",
    "data/flows/Cross-border Physical Flows France 2019.csv",
    "data/flows/Cross-border Physical Flows France 2020.csv",
    "data/flows/Cross-border Physical Flows France 2021.csv",
    "data/flows/Cross-border Physical Flows France 2022.csv",
    "data/flows/Cross-border Physical Flows France 2023.csv",
    "data/flows/Cross-border Physical Flows France 2024.csv",
    "data/flows/Cross-border Physical Flows France 2025.csv",
]

flows_list = []

for path in flow_files:
    df = pd.read_csv(path)
    print(f"Lu : {path}  →  {len(df)} lignes")

    # Colonne MTU (nom variable selon les fichiers)
    if "MTU" in df.columns:
        mtu_col = "MTU"
    elif "MTU (CET/CEST)" in df.columns:
        mtu_col = "MTU (CET/CEST)"
    else:
        print(f"  MTU introuvable, fichier ignoré")
        continue

    df["datetime"] = pd.to_datetime(
        df[mtu_col].astype(str).str.split(" - ").str[0].str.replace(r"\s*\(.*\)", "", regex=True),
        dayfirst=True,
        errors="coerce"
    )

    df["Physical Flow (MW)"] = pd.to_numeric(df["Physical Flow (MW)"], errors="coerce")

    df["from_country"] = df["Out Area"].str.split("|", n=1, regex=False).str[1].map(zone_map)
    df["to_country"]   = df["In Area"].str.split("|",  n=1, regex=False).str[1].map(zone_map)

    # Garder uniquement les interconnexions avec la France
    df = df[(df["from_country"] == "France") | (df["to_country"] == "France")]

    df["partner"] = np.where(
        df["from_country"] == "France",
        df["to_country"],
        df["from_country"]
    )

    df["flow_mw"] = df["Physical Flow (MW)"]
    df["year"]    = df["datetime"].dt.year

    df = (
        df.groupby(["datetime", "year", "from_country", "to_country", "partner"], as_index=False)
        ["flow_mw"].sum()
    )

    flows_list.append(df[["datetime", "year", "from_country", "to_country", "partner", "flow_mw"]])

flows = pd.concat(flows_list, ignore_index=True)
print(f"\nTotal : {len(flows):,} lignes — partenaires : {sorted(flows['partner'].dropna().unique())}")
flows.head()
```

## 3. Capacités journalières (week-ahead)

```python
daily_cap_files = [
    "data/capacities/daily/Forecast transfer capacities week-ahead France 2014.csv",
    "data/capacities/daily/Forecast transfer capacities week-ahead France 2015.csv",
    "data/capacities/daily/Forecast transfer capacities week-ahead France 2016.csv",
    "data/capacities/daily/Forecast transfer capacities week-ahead France 2017.csv",
    "data/capacities/daily/Forecast transfer capacities week-ahead France 2018.csv",
    "data/capacities/daily/Forecast transfer capacities week-ahead France 2019.csv",
    "data/capacities/daily/Forecast transfer capacities week-ahead France 2020.csv",
    "data/capacities/daily/Forecast transfer capacities week-ahead France 2021.csv",
    "data/capacities/daily/Forecast transfer capacities week-ahead France 2022.csv",
    "data/capacities/daily/Forecast transfer capacities week-ahead France 2023.csv",
    "data/capacities/daily/Forecast transfer capacities week-ahead France 2024.csv",
    "data/capacities/daily/Forecast transfer capacities week-ahead France 2025.csv",
]

cap_list = []

for path in daily_cap_files:
    df = pd.read_csv(path)
    print(f"Lu : {path}  →  {len(df)} lignes")

    df["Capacity (MW)"] = pd.to_numeric(df["Capacity (MW)"], errors="coerce")
    df = df.rename(columns={"Capacity (MW)": "capacity_mw"})

    df["from_country"] = df["Out Area"].str.extract(r"\|(.*)")[0].map(zone_map)
    df["to_country"]   = df["In Area"].str.extract(r"\|(.*)")[0].map(zone_map)

    df["day"] = pd.to_datetime(
        df["Day"].str.split(" ").str[1],
        format="%d/%m/%Y"
    )

    df = (
        df.groupby(["day", "from_country", "to_country"], as_index=False)
        ["capacity_mw"].sum()
    )

    cap_list.append(df[["day", "from_country", "to_country", "capacity_mw"]])

daily_capacity = pd.concat(cap_list, ignore_index=True)
print(f"\nTotal : {len(daily_capacity):,} lignes")
daily_capacity.head()
```

```python
prices.to_parquet("data/prices.parquet", index=False)
flows.to_parquet("data/flows.parquet", index=False)
daily_capacity.to_parquet("data/daily_capacity.parquet", index=False)

print("Sauvegardé :")
print(f"  data/prices.parquet        ({len(prices):,} lignes)")
print(f"  data/flows.parquet         ({len(flows):,} lignes)")
print(f"  data/daily_capacity.parquet ({len(daily_capacity):,} lignes)")
```
